```python
zx = np.array([23, 34, 23, 40, 43, 45])
ab = np.array([10, 20, 30, 40, 50, 54])
```

```python
plt.plot(zx, ab, marker = "^", ms = 10, mec = "red", mfc = "yellow")

for i in range(len(zx)):
    plt.text(zx[i], ab[i], f"({zx[i]}, {ab[i]})")

```

```python
plt.plot(zx, ab, marker = "^", ms = 10, mec = "red", mfc = "yellow")

for i in range(len(zx)):
    plt.text(zx[i], ab[i], f"({zx[i]}, {ab[i]})")

plt.xlabel("Employee Age")
plt.ylabel("Employee Sallery")
plt.title("Employee info", color = "red")


# fig = plt.figure(edgecolor= "yellow", facecolor="green")

plt.grid(axis = "both")

# "y"
# "x"
# "both"
```

```python
x = np.array([23, 45, 23, 45, 2, 56, 34, 23, 21])
y = np.array([34, 56, 34, 78, 34, 67, 34, 23, 10])

a = np.array([45, 34, 45, 23, 12])
b = np.array([45, 56, 34, 56, 34])
```

```python
plt.scatter(x, y, color = "red", label = "loss")

# plt.scatter(a, b, color = "blue", label = "prfit")

plt.legend(loc = "upper left")

for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", color = "green")

for i in range(len(a)):
    plt.text(a[i], b[i], f"({a[i]}, {b[i]})", color = "gray")
    plt.xlabel("loss")
```
