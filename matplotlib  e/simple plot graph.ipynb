```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([10, 20, 34, 45, 67, 89, 90])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker = "*", ms = 15)
```

```python
plt.plot(x, y, marker = "<", ms = 15)
```

```python
plt.plot(x, y, marker = ">", ms = 15)
```

```python
plt.plot(x, y, marker = "1", ms = 15)
```

```python
# mfc ---> marker face color
# mec ----> marker edge color

plt.plot(x, y, marker = "*", ms = 15, mec = "red", mfc = "yellow")
```

```python
plt.plot(x, y, marker = ".", ms = 15, mec = "red", mfc = "yellow")
```

```python
plt.plot(x, y, marker = ".", ms = 15, mec = "red", mfc = "yellow",
        linestyle = "dashdot", color = "green", linewidth = 2 )

plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("My Revnew", color = "green", loc = "center")
```

```python
months = np.array([1, 3, 5, 7, 9, 11])
profile = np.array([12, 23, 34, 20, 39, 89])

plt.plot(months, profile, marker = ".", ms = 12, mec = "blue", mfc = "green", 
         linestyle = "dashdot", color = "red", linewidth = 2)

plt.xlabel("This is months")
plt.ylabel("this is profit")

plt.grid()


for i in range(len(months)):
    plt.annotate(xy = (months[i], profile[i]), text = f"({months[i]}, {profile[i]})")
```
