```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xyz = np.array([12, 34, 45, 56, 67, 89, 90])
```

```python
abc = np.array([12, 34, 23, 34, 45, 34, 23])
```

```python
plt.plot(xyz, abc)
```

```python
plt.plot(xyz, abc, marker = ".")
```

```python
plt.plot(xyz, abc, marker = "*")
```

```python
plt.plot(xyz, abc, marker = "*", ms = 16)
```

```python
# marker color:
#     mec => marker edge color
#     mfc -> maker face color

plt.plot(xyz, abc, marker = "*", ms = 16, mec = "red")
```

```python
# marker color:
#     mec => marker edge color
#     mfc -> maker face color

plt.plot(xyz, abc, marker = "*", ms = 16, mec = "red", mfc = "yellow")
```

```python
# line 

plt.plot(xyz, abc, marker = "*", ms = 16, mec = "red", mfc = "yellow", color = "green")
```

```python
# line 

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 5)
```

```python
# line 

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dotted")
```

```python
# line 

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashed")
```

```python
# line 

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashdot")
```

```python
# line 

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dotted")
```

```python

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashdot")
plt.xlabel("this is x axis xyz")
plt.ylabel("this is y axis ABC")
```

```python

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashdot")
plt.xlabel("this is x axis xyz")
plt.ylabel("this is y axis ABC")
plt.title("this is title", color = "red")
```

```python

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashdot")
plt.xlabel("this is x axis xyz")
plt.ylabel("this is y axis ABC")
plt.title("this is title", color = "red")
plt.grid()
```

```python

plt.plot(xyz, abc, 
         marker = "*", ms = 16, mec = "red", mfc = "yellow", 
         color = "green", linewidth = 2, linestyle = "dashdot")
plt.xlabel("this is x axis xyz")
plt.ylabel("this is y axis ABC")
plt.title("this is title", color = "red")
plt.grid(color = "green", linestyle = "dashdot", linewidth = 0.4)
```
