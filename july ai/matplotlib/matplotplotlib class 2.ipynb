```python
xyz = np.array([12, 34, 45, 67, 9, 3, 45, 4])
abc = np.array([34, 45, 56, 67, 78, 45, 34, 23])

clr = ["red", "pink", "yellow", "blue", "green", "black", "orange", "lightgreen"]

size = [100, 300, 400, 200, 500, 300, 600, 488]

plt.scatter(xyz, abc, color = clr, s = size)

for i in range(len(xyz)):
    plt.annotate(xy = (xyz[i], abc[i]), text = f"({xyz[i]}, {abc[i]})")
```

```python
xyz = np.array([12, 34, 45, 67, 9, 3, 45, 4])
abc = np.array([34, 45, 56, 67, 78, 45, 34, 23])

clr = ["red", "pink", "yellow", "blue", "green", "black", "orange", "lightgreen"]

size = [100, 300, 400, 200, 500, 300, 600, 488]

plt.scatter(xyz, abc, color = clr, s = size, label = "data point")

for i in range(len(xyz)):
    plt.annotate(xy = (xyz[i], abc[i]), text = f"({xyz[i]}, {abc[i]})")

plt.xlabel("thissis x label")
plt.ylabel("thissis y label")
plt.title("thiss is title")
plt.legend()
```

```python
xyz = np.array([12, 34, 45, 67, 9, 3, 45, 4])
abc = np.array([34, 45, 56, 67, 78, 45, 34, 23])

clr = ["red", "pink", "yellow", "blue", "green", "black", "orange", "lightgreen"]

size = [100, 300, 400, 200, 500, 300, 600, 488]

plt.scatter(xyz, abc, color = clr, s = size, label = "data point")

plt.plot(xyz, abc, color = "red", marker = ".", ms = 13, mec = "red", 
         mfc = "yellow", linestyle = "dashdot", label = "trend line")

for i in range(len(xyz)):
    plt.annotate(xy = (xyz[i], abc[i]), text = f"({xyz[i]}, {abc[i]})")

plt.xlabel("thissis x label")
plt.ylabel("thissis y label")
plt.title("thiss is title")
plt.legend()
```

```python
month = np.array(["Jan", "Feb", "March", "April", "May"])
profit = np.array([10, 12, 5, 8, 6])

clr = ["red", "blue", "lightgreen", "pink", "green"]

sizes = [0.7, 0.8, 0.6, 0.5, 0.9]

plt.barh(month, profit, color = clr)

plt.xlabel("Months")
plt.ylabel("profit")
plt.title("revnue")
```

```python
kg = np.array([50, 30, 40, 20, 40])

lb = ["Apple", "Bananas", "Grapes", "Orange", "Melon"]

plt.pie(kg, labels = lb)
```

```python
kg = np.array([50, 30, 40, 20, 40])

lb = ["Apple", "Bananas", "Grapes", "Orange", "Melon"]
clr = ["red", "yellow", "green", "orange", "lightgreen"]

plt.pie(kg, labels = lb, colors = clr)
```

```python
kg = np.array([50, 30, 40, 20, 40])

lb = ["Apple", "Bananas", "Grapes", "Orange", "Melon"]
clr = ["red", "yellow", "green", "orange", "lightgreen"]

explod = [0.1, 0, 0, 0, 0]

plt.pie(kg, labels = lb, colors = clr, explode = explod)
```
